```python
#Import data
#mengambil data dan di simpan pada variable df 
#df = pd.read_csv(url)
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/rizalespe/Dataset-Sentimen-Analisis-Bahasa-Indonesia/master/dataset_tweet_sentiment_opini_film.csv', index_col = False)
df.shape
```

```python
df
```

```python
def preprocess (doc):
    url = re.compile(r"https://\S+|www\.+S|&amp\S+")
    clean = url.sub(r"",doc)
    html = re.compile(r"<.*?>")
    clean =  html.sub(r"",clean)
    mention = re.compile(r"@\S+|#\S+")
    clean = mention.sub(r"",clean)
    clean = re.sub('[0-9]+', '', clean)
    clean = re.sub('#',"",clean)
    return clean
```

```python
df['cleanText']=df['Text Tweet'].apply(lambda x: preprocess(x.lower()))
```
